### 2.2 Add more info


```python
# Extract the year
d_excel['Year'] = d_excel['Fecha'].dt.year

# Extract the Quarter
d_excel['Quarter'] = d_excel['Fecha'].dt.quarter

# Extract the Month
d_excel['Month'] = d_excel['Fecha'].dt.month

# Extract the week
d_excel['Week'] = d_excel['Fecha'].dt.week

# Extract the weekday
d_excel['Weekday'] = d_excel['Fecha'].dt.weekday

# Extract the Day
d_excel['Day'] = d_excel['Fecha'].dt.day

d_excel.head(15)
```

### 2.3 Reindex

```python
# Single Index


d_excel = d_excel.set_index('Fecha')
d_excel.head(15)
```

```python
# Multiple index
d_excel2 = d_excel
d_excel2 = d_excel.set_index(['Year','Quarter'], append = True )
d_excel2.head(15)
```
