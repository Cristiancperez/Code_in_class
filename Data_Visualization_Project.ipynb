# **DATA VISUALIZATION PROJECT**


# **1.VARIABLE**


## **1.1. Name**

Exports from Colombia's free trade zones


## **1.2. Period and Frequency**

From 2007 to 2021, with a monthly frequency.


## **1.3. Source**
[DANE](https://www.dane.gov.co/index.php/estadisticas-por-tema/comercio-internacional/zonas-francas)



# **2. GOAL OF DE DV**


## **2.1. Story**

Free trade zones are geographically delimited areas within the national territory, they have special tax, customs and foreign trade regulations, in 2021 the income tax for legal entities was 31% at a general level, these zones have a special rate of 20%.

However, the tax reform proposes to repeal this special rate for free zones that do not export and it will be mandatory for free zones to have an internationalization plan and a minimum export ceiling.
 image.png

 




## **2.2. Audience**

This project is aimed at those interested in seeing the contribution of free trade zones to Colombia's exports, such as foreign or national investors, academics and why not? ordinary people interested in understanding why the current government wants to encourage free trade zones to export more.

## **2.3. Justify the DV type**

To show how the participation of the free trade zones in the total exports of the country has varied, by means of a pie chart where the beginning and the end of the chosen period can be found and compared.
